```python
import pandas as pd
import numpy as np
from urllib import request
url="https://www.nddb.coop/information/stats/milkprodstate"
raw=request.urlopen(url).read().decode('utf8')
df=pd.read_html(raw,header=1,index_col=0)
milk=df[0].T
milk.shape
```

```python
ut=milk.iloc[:,-8:-1]
ut=ut.fillna(0)
ut=ut.astype('int64')
ut
```

```python
st=milk.iloc[:,-37:-8]
st['Telangana']=st['Telangana'].replace(to_replace='-',value=0)
st=st.astype('int64')
st
```
